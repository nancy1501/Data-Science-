```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

#DROP UNNAMED COLUMN

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
#Data Transformation
df["WklyStudyHours"] = df["WklyStudyHours"]
df.head()
```

# Gender Distribution

```python
plt.figure(figsize = (5, 5))
ax = sns.countplot(data = df, x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

#From the above chart we have analysed that:
#the number of females in the datais more than the number of males

#now we analyse parent education impact of different scores

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize = (5, 5))
sns.heatmap(gb, annot = True)
plt.title("Relationship Between Parent's Education and Student's Score")
plt.show()
```

#from the above chart we have concluded that the education of parent have a good impact on their kids score

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize = (5, 5))
sns.heatmap(gb1, annot = True)
plt.title("Relationship Between Parent's Marital Status and Student's Score")
plt.show()
```

```python
gb2 = df.groupby("NrSiblings").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb2)
```

```python
plt.figure(figsize = (5, 5))
sns.heatmap(gb2, annot = True)
plt.title("Relationship Between NrSiblings and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that the NrSiblings have negligible any impact on their kids score
```

#Distribution of Ethnic Groups

```python
import matplotlib.pyplot as plt

# Count students in each ethnic group
groupA = df[df['EthnicGroup'] == "group A"].shape[0]  # Number of rows in group A
groupB = df[df['EthnicGroup'] == "group B"].shape[0]  # Number of rows in group B
groupC = df[df['EthnicGroup'] == "group C"].shape[0]  # Number of rows in group C
groupD = df[df['EthnicGroup'] == "group D"].shape[0]  # Number of rows in group D
groupE = df[df['EthnicGroup'] == "group E"].shape[0]  # Number of rows in group E

# Store counts in a list
counts = [groupA, groupB, groupC, groupD, groupE]

# Labels for the pie chart
labels = ["Group A", "Group B", "Group C", "Group D", "Group E"]

# Plot the pie chart
plt.pie(counts, labels=labels, autopct='%1.2f%%')  # autopct adds percentages
plt.title("Distribution of Students by Ethnic Group")
plt.show()
```
